**Задание**:реализовать генетический алгоритм, решающий нелинейное диофантово уравнение.
            Использовать заданные виды стратегии этапов алгоритма.
            Необходимо самостоятельно экспериментально подобрать наиболее подходящие размеры популяций,
            число особей производящих потомков, необходимые вероятности, необходимые размеры замещаемых частей популяции и т.д. 

**НП**:          Случайная в (-200,200) 

**Селекция**:    Турнирная 

**Cкрещивание**: Пропорциональное вероятностное 

**Мутация**:     Каждый бит всех потомков мутирует с некоторой вероятностью p 

**Замещение**:   Случайно пропорциональным образом создать новую популяцию из предков и потомков 

```python
class Function:
    
    def __init__(self,result,powU,powW,powX,powY,powZ,
                         powU1,powW1,powX1,powY1,powZ1,
                         powU2,powW2,powX2,powY2,powZ2,
                         powU3,powW3,powX3,powY3,powZ3,
                         powU4,powW4,powX4,powY4,powZ4):
        self.result=result
        self.powU=powU
        self.powW=powW
        self.powX=powX
        self.powY=powY
        self.powZ=powZ
        self.powU1=powU1
        self.powW1=powW1
        self.powX1=powX1
        self.powY1=powY1
        self.powZ1=powZ1
        self.powU2=powU2
        self.powW2=powW2
        self.powX2=powX2
        self.powY2=powY2
        self.powZ2=powZ2
        self.powU3=powU3
        self.powW3=powW3
        self.powX3=powX3
        self.powY3=powY3
        self.powZ3=powZ3
        self.powU4=powU4
        self.powW4=powW4
        self.powX4=powX4
        self.powY4=powY4
        self.powZ4=powZ4
    
    def getResult(self):
        return self.result
             
    def calculateFunction(self,u,w,x,y,z):
        return u**self.powU*w**self.powW*x**self.powX*y**self.powY*z**self.powZ+u**self.powU1*w**self.powW1*x**self.powX1*y**self.powY1*z**self.powZ1+u**self.powU2*w**self.powW2*x**self.powX2*y**self.powY2*z**self.powZ2+u**self.powU3*w**self.powW3*x**self.powX3*y**self.powY3*z**self.powZ3+u**self.powU4*w**self.powW4*x**self.powX4*y**self.powY4*z**self.powZ4
    
    def __str__(self):
        return "u**%d*w**%d*x**%d*y**%d*z**%d+u**%d*w**%d*x**%d*y**%d*z**%d+u**%d*w**%d*x**%d*y**%d*z**%d+u**%d*w**%d*x**%d*y**%d*z**%d+u**%d*w**%d*x**%d*y**%d*z**%d"%(self.powU,self.powW,self.powX,self.powY,self.powZ,self.powU1,self.powW1,self.powX1,self.powY1,self.powZ1,self.powU2,self.powW2,self.powX2,self.powY2,self.powZ2,self.powU3,self.powW3,self.powX3,self.powY3,self.powZ3,self.powU4,self.powW4,self.powX4,self.powY4,self.powZ4)
```

```python
import random

FROM=-200
TO=200

class Individual:
    
    def __init__(self,function):
        self.x=random.randint(FROM,TO)
        self.y=random.randint(FROM,TO)
        self.z=random.randint(FROM,TO)
        self.u=random.randint(FROM,TO)
        self.w=random.randint(FROM,TO)
        self.function=function
        self.probability=0
        self.fa=abs(function.calculateFunction(self.u,self.w,self.x,self.y,self.z-function.getResult()))

    def getX(self):
        return self.x
    
    def getY(self):
        return self.y
    
    def getZ(self):
        return self.z
    
    def getU(self):
        return self.u
    
    def getW(self):
        return self.w
                    
    def getFa(self):
        return self.fa
    
    def getProbability(self):
        return self.probability
    
    def setX(self,x):
        self.x=x
        
    def setY(self,y):
        self.y=y
        
    def setZ(self,z):
        self.z=z
        
    def setU(self,u):
        self.u=u
        
    def setW(self,w):
        self.w=w
    
    def setFa(self):
        self.fa=abs(self.function.calculateFunction(self.u,self.w,self.x,self.y,self.z-self.function.getResult()))
    
    def setProbability(self,p):
        self.probability=p
        
    def increaseProbability(self,i):
        self.probability=self.probability*i
                    
    def __str__(self):
        return "(%d,%d,%d,%d,%d); fa=%d" % (self.u,self.w,self.x,self.y,self.z,self.fa)
```

```python
import sys
import timeit
import time 

POPULATION_SIZE=100
SELECTION=40
NUMBER_OF_SELECTION=2
MAX_REPEAT=100
PROBABILITY_OF_MUTATION=0.15
PROBABILITY_OF_MUTATION_WHILE_NUCLEAR_BLAST=0.75


class Population:
    
    def __init__(self,function):
        self.function=function
        self.population=[Individual(self.function) for i in range(POPULATION_SIZE)]
        self.numberOfGeneration=0
        self.mutation=PROBABILITY_OF_MUTATION
    
    def calculateProbabilityForElement(self,element,elements):
        probability=1/element.getFa()
        denominator=0
        for i in range(len(elements)):
            denominator=denominator+1/elements[i].getFa()
        if not(denominator==0):
            probability=probability/denominator
        return probability
    
    def addprobability(self,element,elements):
        probability=elements[0].getProbability()
        for i in range(1,len(elements)):
            probability=probability+elements[i].getProbability()
        return probability
    
    def crossbreeding(self,firstparant,secondparent):
        kid=Individual(self.function)
        firstparant.setProbability(self.calculateProbabilityForElement(firstparant,[firstparant,secondparent]))
        firstparant.increaseProbability(10**5)
        secondparent.setProbability(self.calculateProbabilityForElement(secondparent,[firstparant,secondparent]))
        secondparent.setProbability(firstparant.getProbability()+secondparent.getProbability())
        secondparent.increaseProbability(10**10)
        parents=[firstparant,secondparent]
        #u
        parents.sort(key=lambda x:x.u)
        randomSelect=random.uniform(0,10**10)
        if randomSelect<parents[0].getProbability():
            kid.setU(parents[0].getU())
        else:
            kid.setU(parents[1].getU())
        #w
        parents.sort(key=lambda x:x.w)
        randomSelect=random.uniform(0,10**10)
        if randomSelect<parents[0].getProbability():
            kid.setW(parents[0].getW())
        else:
            kid.setW(parents[1].getW())
        #x
        parents.sort(key=lambda x:x.x)
        randomSelect=random.uniform(0,10**10)
        if randomSelect<parents[0].getProbability():
            kid.setX(parents[0].getX())
        else:
            kid.setX(parents[1].getX())
        #y
        parents.sort(key=lambda x:x.y)
        randomSelect=random.uniform(0,10**10)
        if randomSelect<parents[0].getProbability():
            kid.setY(parents[0].getY())
        else:
            kid.setY(parents[1].getY())     
        #z
        parents.sort(key=lambda x:x.z)
        randomSelect=random.uniform(0,10**10)
        if randomSelect<parents[0].getProbability():
            kid.setZ(parents[0].getZ())
        else:
            kid.setZ(parents[1].getZ())   
        kid.setFa()
        return kid
    
    def mutate(self,individual):
        issomethingchange = False
        temp = random.random()
        if temp<=self.mutation:
            issomethingchange=True
            individual.setU(random.randint(FROM,TO))
        temp = random.random()
        if temp<=self.mutation:
            issomethingchange=True
            individual.setW(random.randint(FROM,TO))
        temp = random.random()
        if temp<=self.mutation:
            issomethingchange=True
            individual.setX(random.randint(FROM,TO))
        temp = random.random()
        if temp<=self.mutation:
            issomethingchange=True
            individual.setY(random.randint(FROM,TO))
        temp = random.random()
        if temp<=self.mutation:
            issomethingchange=True
            individual.setZ(random.randint(FROM,TO))
        if issomethingchange:
            individual.setFa()
        
    def replacegeneration(self,allpopulation):
        newgeneration=[]
        while len(newgeneration)<=POPULATION_SIZE:
            for i in range(len(allpopulation)):
                allpopulation[i].setProbability(self.calculateProbabilityForElement(allpopulation[i],allpopulation))
                allpopulation[i].increaseProbability(10**3)
                if i>0:
                    allpopulation[i].setProbability(self.addprobability(allpopulation[i],allpopulation[:i]))
            temp = random.uniform(0,allpopulation[len(allpopulation)-1].getProbability())
            isGet=False
            for j in range(1,len(allpopulation)):
                if allpopulation[j-1].getProbability()<temp and temp<=allpopulation[j].getProbability():
                    newgeneration.append(allpopulation.pop(j))
                    ifGet=True
            if not isGet:
                newgeneration.append(allpopulation.pop(0))
        newgeneration.sort(key=lambda x: x.fa)
        return newgeneration

    def checkFA(self,elements):
        for i in range(len(elements)):
            if elements[i].getFa()==0:
                return True
        return False
    
    def getNewPopulation(self):
        #Селекця
        selectedfromoldgeneration=[]
        self.population.sort(key=lambda x: x.fa)
        for i in range(SELECTION):
            selectedfromoldgeneration.append(self.population[i])
        #Скрещивание
        childrens=[]
        for start in range(len(selectedfromoldgeneration)-1):
            for number in range(NUMBER_OF_SELECTION):
                selectedIndex=random.randint(0,len(selectedfromoldgeneration)-1)
                while selectedIndex==start:
                    selectedIndex=random.randint(0,len(selectedfromoldgeneration)-1)
                childrens.append(self.crossbreeding(selectedfromoldgeneration[start],selectedfromoldgeneration[selectedIndex]))
        #Мутация
        for j in range(len(childrens)):
            self.mutate(childrens[j])
        if self.checkFA(childrens):
            return [True,childrens]
        #Замещение
#         childrens.sort(key=lambda x: x.fa)
#         return childrens[:POPULATION_SIZE]
        allindividuals=[]
        for p in range(len(self.population)):
            allindividuals.append(self.population[p])
        for c in range(len(childrens)):
            allindividuals.append(childrens[c])
        return [False,self.replacegeneration(allindividuals)]
        
    def run(self):
        t = time.time()
        repeat=0
        isNuclearBlast=False
        self.population.sort(key=lambda x: x.fa)
        self.bestfromall=self.population[0]
        self.bestfromthisgeneration=self.population[0]
        print("Покаление номер ",self.numberOfGeneration,end="")
        print(f" Лучший из начального поколения: {self.bestfromall}")
        while 1:
            if self.bestfromthisgeneration.getFa()==0:
                break
            if repeat>MAX_REPEAT:
                isNuclearBlast=True
                self.mutation=PROBABILITY_OF_MUTATION_WHILE_NUCLEAR_BLAST
            print("Поколение номер ",self.numberOfGeneration+1,end="")
            [flag,newpopulation]=self.getNewPopulation()
            
            self.population.clear()
            for j in range(len(newpopulation)):
                self.population.append(newpopulation[j])
            self.population.sort(key=lambda x: x.fa)
            self.bestfromthisgeneration=self.population[0]
            if flag==True:
                self.bestfromall=self.bestfromthisgeneration
                break
            if self.bestfromthisgeneration.getFa()<=self.bestfromall.getFa():
                self.bestfromall=self.bestfromthisgeneration
                repeat=0
            else:
                repeat=repeat+1
            print(f" Лучший из всех поколений: {self.bestfromall}",
                  f" Лучший из {self.numberOfGeneration+1} поколений: {self.bestfromthisgeneration}")
            if isNuclearBlast:
                self.mutation=PROBABILITY_OF_MUTATION
                repeat=0
            self.numberOfGeneration=self.numberOfGeneration+1
        print(f" Решение {self.bestfromall} и количество поколений {self.numberOfGeneration+1}")
        print(time.time()-t)
```
